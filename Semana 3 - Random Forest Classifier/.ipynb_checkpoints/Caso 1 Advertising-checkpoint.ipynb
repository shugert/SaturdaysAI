```python
dataset = pd.read_csv('advertising.csv')
dataset.head()
```

<h3>Extraer variable: Datetime</h3>

```python
dataset['Timestamp'] = pd.to_datetime(dataset['Timestamp'])
dataset['Timestamp']
dataset['Year'] = dataset['Timestamp'].dt.year
dataset['Month'] = dataset['Timestamp'].dt.month
dataset['Day'] = dataset['Timestamp'].dt.day
dataset['Hour'] = dataset['Timestamp'].dt.hour
dataset['Weekday'] = dataset['Timestamp'].dt.dayofweek
dataset = dataset.drop(['Timestamp'], axis=1)
dataset.head(10)
```

<h3>Relación entre variables</h3>

```python
# Relación entre variables númericas.
sns.pairplot(dataset, hue = 'Clicked on Ad', 
             vars = ['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage'], palette = 'Greens_r')
```

```python
# como ya creamos valores para mes, día y hora, vamos a borrar la columna año que correspondia al datetime
dataset = dataset.drop(['Year'], axis=1)
```

```python
# Mapa de correlacion con las nuevas variables
fig = plt.figure(figsize = (12,10))
sns.heatmap(dataset.corr(), cmap='Greens', annot = True)
```

```python
# sanity check!
dataset.head()
```

<h3>Construir un modelo de Regresión Logistica</h3>

```python
# primero asignamos los valores de x y de y, siendo este último nuestro valor que queremos predecir.
X = dataset.iloc[:,[0,1,2,3,6,9,10,11,12]].values
y = dataset.iloc[:,8].values
```

```python
# particionamos el dataset en dos, uno para entrenamiento y el otro para comprobar nuestro modelo.

# tenemos que importar una nueva libreria para hacer el split. 
# dejaremos 80% para el training y 20% para el test.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Escalar los datos
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
X_train = standardScaler.fit_transform(X_train)
X_test = standardScaler.transform(X_test)
```

```python
# iniciar y ajustar el modelo de regresión logistica en el dataset de entrenamiento.

# primero importamos la libreria de Regresion Logistica de SkLearn
from sklearn.linear_model import LogisticRegression
log_rg = LogisticRegression()
log_rg.fit(X_train, y_train)

# Predicción
y_log_rg = log_rg.predict(X_test)
```

<h3>Evaluar Resultados</h3>

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_log_rg)
sns.heatmap(cm,annot=True,fmt='3.0f',cmap="Greens")
plt.title('Confusion matrix para Regresión Logistica', y=1.05, size=15)
```

```python
# Reporte de Clasificación
from sklearn.metrics import classification_report
cr = classification_report(y_test, y_log_rg)
print(cr)
```

<h3>Probemos ahora con un modelo de Naive Bayes</h3>

```python
# Iniciar y ajustar el modelo de Naive Bayes al dataset de entrenamiento.
from sklearn.naive_bayes import GaussianNB
naive_b = GaussianNB()
naive_b.fit(X_train, y_train)

#Predicción
y_naive = naive_b.predict(X_test)
```

<h3>Evaluar Resultados</h3>

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
naive_cm = confusion_matrix(y_test, y_naive)
sns.heatmap(naive_cm,annot=True,fmt='3.0f',cmap="Blues")
plt.title('Confusion matrix para Naive Bayes', y=1.05, size=15)
```

```python
# Reporte de Clasificación
from sklearn.metrics import classification_report
naive_cr = classification_report(y_test, y_naive)
print(naive_cr)
```

<h3>Modelo Random Forest</h3>

```python
# Random Forest en dataset de entrenamiento
from sklearn.ensemble import RandomForestClassifier
randm_frst = RandomForestClassifier()
randm_frst.fit(X_train, y_train)

#Predicción
y_frst = randm_frst.predict(X_test)
```

<h3>Evaluar Resultados</h3>

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
frst_cm = confusion_matrix(y_test, y_frst)
sns.heatmap(frst_cm,annot=True,fmt='3.0f',cmap="Reds")
plt.title('Confusion matrix para Random Forest', y=1.05, size=15)
```

```python
# Reporte de Clasificación
from sklearn.metrics import classification_report
frst_cr = classification_report(y_test, y_frst)
print(frst_cr)
```

<h3>Modelo K-Nearest Neighbors</h3>

```python
#K-Nearest Neighbors en dataset de entrenamiento
from sklearn.neighbors import KNeighborsClassifier
kneighbors = KNeighborsClassifier()
kneighbors.fit(X_train, y_train)

#Predicción
y_knn = kneighbors.predict(X_test)
```

<h3>Evaluar Resultados</h3>

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
knn_cm = confusion_matrix(y_test, y_knn)
sns.heatmap(knn_cm,annot=True,fmt='3.0f',cmap="mako")
plt.title('Confusion matrix para K-Nearest Neighbors', y=1.05, size=15)
```

```python
# Reporte de Clasificación
from sklearn.metrics import classification_report
knn_cr = classification_report(y_test, y_knn)
print(knn_cr)
```

<h3>Comparación de modelos en base a f1 score</h3>

```python
from sklearn.metrics import f1_score
f1_log = f1_score(y_test, y_log_rg)
f1_naive = f1_score(y_test, y_naive)
f1_frst = f1_score(y_test, y_frst)
f1_knn = f1_score(y_test, y_knn)
```

```python
from pandas import DataFrame
scores = {'Model':  ['Regresión Logistica','Naive_Bayes', 'Random Forest', 'KNN'], 
          'f1 score': [f1_log, f1_naive, f1_frst, f1_knn]}
f1_scores = DataFrame (scores, columns = ['Model','f1 score'])
f1_scores
```

```python
sns.barplot(x="Model", y="f1 score", data=f1_scores, palette="Greens_r")
```

<h3>Fine-tuning de modelos</h3>

```python
# Escoger hiperparametros para modelo Random Forest
from sklearn.model_selection import GridSearchCV
param_frst = [{"n_estimators": [10,100,200,300,500], "criterion": ["gini", "entropy"]}]
grid_search_frst = GridSearchCV(estimator=randm_frst,
                          param_grid=param_frst,
                          scoring = 'accuracy',
                          cv=10)
grid_search_frst = grid_search_frst.fit(X_train, y_train)
```

```python
# Calcular accuracy para Random Forest 
best_acc_frst = grid_search_frst.best_score_
best_acc_frst
```

```python
# Calcular los mejores parametros para el modelo Random Forest
best_params_frst = grid_search_frst.best_params_
best_params_frst
```

```python
#Escoger hyperparameters para K-Nearest
from sklearn.model_selection import GridSearchCV
param_knn = [{"n_neighbors": range(1,10), "weights": ["uniform", "distance"]}]
grid_search_knn = GridSearchCV(estimator=kneighbors,
                          param_grid=param_knn,
                          scoring = 'accuracy',
                          cv=10)
grid_search_knn = grid_search_knn.fit(X_train, y_train)
```

```python
#Calcular accuracy para K-Nearest Neighbors 
best_acc_knn = grid_search_knn.best_score_
best_acc_knn
```

```python
# Calcular los mejores parametros para el modelo K-Nearest Neighbors 
best_params_knn = grid_search_knn.best_params_
best_params_knn
```

<h3>Reconstruir modelos con hyperparametros</h3>

```python
# Iniciar y ajustar Random Forest en dataset de entrenamiento con los hiperparametros
from sklearn.ensemble import RandomForestClassifier
randm_frst_imp = RandomForestClassifier(n_estimators=100, criterion='gini')
randm_frst_imp.fit(X_train, y_train)
#Prediccion
y_frst_imp = randm_frst_imp.predict(X_test)
```

```python
# Iniciar y ajustar K-Nearest Neighbors en dataset de entrenamiento con los hiperparametros
from sklearn.neighbors import KNeighborsClassifier
kneighbors_imp = KNeighborsClassifier(n_neighbors=5, weights= 'uniform')
kneighbors_imp.fit(X_train, y_train)
#Prediccion
y_knn_imp = kneighbors_imp.predict(X_test)
```

```python
#Recalcular f1 score
f1_frst_imp = f1_score(y_test, y_frst_imp)
f1_knn_imp = f1_score(y_test, y_knn_imp)
```

```python
scores_imp = {'Model':  ['Regresión Logistica','Naive_Bayes', 'Random Forest', 'KNN'], 
          'f1 score': [f1_log, f1_naive, f1_frst_imp, f1_knn_imp]}
f1_scores_imp = DataFrame (scores_imp, columns = ['Model','f1 score'])
f1_scores_imp.sort_values(by=['f1 score'], ascending=False)
```

```python
#Confusion matrix para el mejor modelo
frst_imp_cm = confusion_matrix(y_test, y_log_rg)
sns.heatmap(frst_imp_cm,annot=True,fmt='3.0f',cmap="PuBu_r")
plt.title('Confusion matrix para Regresión Logistica', y=1.05, size=15)
```

<h3>Area Bajo la Curva ROC</h3>

```python
from sklearn.metrics import roc_auc_score
lr_auc = roc_auc_score(y_test, log_rg.predict(X_test))
rf_roc_auc = roc_auc_score(y_test, randm_frst_imp.predict(X_test))
```

```python
# Crear ROC Curve
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, log_rg.predict_proba(X_test)[:,1])
rf_fpr, rf_tpr, rf_thresholds = roc_curve(y_test, randm_frst_imp.predict_proba(X_test)[:,1])

plt.figure()
# Plot Random Forest ROC
plt.plot(rf_fpr, rf_tpr, label='Random Forest (area = %0.2f)' % rf_roc_auc)
# Plot Logistic Regression ROC
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % lr_auc)
# Plot Base Rate ROC
plt.plot([0,1], [0,1],label='Base Rate')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.10])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```

<h3>Resultados</h3>

El modelo de Regresión Logistica tuvo el mejor performance en comparación con Random Forest, Naive Bayes and K-Nearest Neighbors.

```python
print(classification_report(y_test, y_frst_imp))
```

El modelo tiene un muy alto f1 score 0.97, que quiere decir que los valores predecidos son are 96% confiables.
El modelo solo tiene 4 puntos mal predecidos en la matriz de confusión.
